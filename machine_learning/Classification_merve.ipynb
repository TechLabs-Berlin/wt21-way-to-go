# Clustering

```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,)

kmeans.fit(y)
```

```python
# The lowest SSE value
kmeans.inertia_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=50,
    max_iter=500,)

kmeans.fit(y)
```

```python
print('The lowest SSE value', kmeans.inertia_)
print('Final locations of the centroid', kmeans.cluster_centers_)
print('The number of iterations required to converge', kmeans.n_iter_)
```

```python
label = kmeans.labels_
label
```

```python
# Elbow method

kmeans_kwargs = {
    "init": "random",
    "n_init": 12,
    "max_iter": 500,
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 12):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(y)
    sse.append(kmeans.inertia_)
    
# Plotting
plt.style.use("fivethirtyeight")
plt.plot(range(1, 12), sse)
plt.xticks(range(1, 12))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
# Determining the elbow point in the SSE 

kl = KneeLocator(
    range(1, 12), sse, curve="convex", direction="decreasing"
)

kl.elbow
```

```python
# Silhouette coefficient

# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 12):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(y)
    score = silhouette_score(y, kmeans.labels_)
    silhouette_coefficients.append(score)

# Plotting
plt.style.use("fivethirtyeight")
plt.plot(range(2, 12), silhouette_coefficients)
plt.xticks(range(2, 12))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```
